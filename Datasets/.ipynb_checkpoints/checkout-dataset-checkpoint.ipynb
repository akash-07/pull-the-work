```python

date1 = '2017-11-16T16:05:16Z'
date2 = '2017-12-05T18:07:27Z'
import re
def getDays(date1, date2):
    vals1 = re.findall(r'[\d]+', date1)
    vals2 = re.findall(r'[\d]+', date2)
    nums1 = [int(x) for x in vals1]
    nums2 = [int(x) for x in vals2]
    days = (nums2[0] - nums1[0])*365 + (nums2[1] - nums1[1]) * 30 + (nums2[2] - nums1[2])
    return days
```

```python
created_dates = df['created_at'].values
closed_dates = df['closed_at'].values
```

```python
closed_dates
```

```python
closing_ages = [getDays(date1, date2) for date1, date2 in zip(created_dates, closed_dates)]
```

```python
closing_ages
```

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(closing_ages,20)
plt.xlabel('days')
plt.ylabel('number of issues')
plt.title('time taken to close issues')
```

```python
address_dates = df['address_date'].values
address_time = [getDays(date1, date2) for date1, date2 in zip(created_dates,address_dates)]
```

```python
address_time
```

```python
plt.hist(address_time, 25)
plt.xlabel('days')
plt.ylabel('number of issues')
plt.title('time for first contributor/member comment')
```

```python
plt.scatter(address_time, closed_dates)
plt.xlabel('first address time in days')
plt.ylabel('closed time in days')
plt.title('close time vs address time')
```
